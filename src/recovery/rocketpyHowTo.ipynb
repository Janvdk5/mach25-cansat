# Introduction to RocketPy

RocketPy is a powerful and easy-to-use trajectory simulation tool for High-Power Rocketry (HPR). 
It allows users to simulate the flight of a rocket, taking into account various factors such as 
aerodynamics, weather conditions, and motor performance. RocketPy is designed to be user-friendly 
and provides detailed outputs that can be used for analysis and optimization of rocket designs.

## Key Features of RocketPy
- **Trajectory Simulation**: Simulate the flight path of a rocket from launch to landing.
- **Aerodynamic Analysis**: Calculate aerodynamic forces and moments acting on the rocket.
- **Weather Conditions**: Incorporate real-time weather data into simulations.
- **Motor Performance**: Model the thrust curve of rocket motors.
- **Optimization**: Optimize rocket design parameters for improved performance.
